```python
# === Dekripsi Related RSA (Polynomial GCD) ===

import json
import sympy as sp
from pathlib import Path

# --- Load file JSON ---
file_path = "franklin.json"  # pastikan file ada di folder kerja / Colab
data = json.loads(Path(file_path).read_text())

# Konversi hex -> int
def hx(x):
    x = x.lower()
    if x.startswith("0x"):
        return int(x, 16)
    return int(x, 16)

n = hx(data["n"])
e = hx(data["e"])
a = hx(data["a"])
b = hx(data["b"])
c1 = hx(data["c1"])
c2 = hx(data["c2"])

# --- Bangun polinomial ---
x = sp.symbols('x')
f = sp.expand((a*x + b)**e - c2)
g = sp.expand(x**e - c1)

F = sp.Poly(f, x, domain=sp.ZZ)
G = sp.Poly(g, x, domain=sp.ZZ)

# --- GCD ---
H = sp.gcd(F, G)
print(f"GCD degree: {sp.degree(H)}")
coeffs = H.all_coeffs()
print("Koefisien GCD:", coeffs)

# --- Cari m1 ---
if sp.degree(H) == 1:
    A, B = int(coeffs[0]), int(coeffs[1])
    if B % A == 0:
        m1 = (-B) // A
    else:
        A_inv = pow(A, -1, n)
        m1 = (-B * A_inv) % n
    m1 = m1 % n
    m2 = (a * m1 + b) % n

    # Konversi ke teks
    def int_to_bytes(i):
        h = hex(i)[2:]
        if len(h) % 2:
            h = "0" + h
        return bytes.fromhex(h)

    m1_bytes = int_to_bytes(m1)
    try:
        plaintext = m1_bytes.decode('utf-8', errors='ignore')
    except:
        plaintext = m1_bytes.hex()

    print("\n=== HASIL DEKRIPSI ===")
    print("m1 =", m1)
    print("Plaintext m1 =", plaintext)
    print("FLAG => RTRTNI25{" + plaintext + "}")
else:
    print("GCD bukan linear, hasil tidak bisa langsung diekstrak.")
```
